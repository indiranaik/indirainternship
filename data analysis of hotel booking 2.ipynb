# data analysation and visualization for cleaned data

```python
canceled_percentile = df['is_canceled'].value_counts(normalize=True)
print(canceled_percentile)
```

```python
canceled_percentile = df['is_canceled'].value_counts(normalize=True)
print(canceled_percentile)

plt.figure(figsize=(5,4))
plt.title('reservation status count')
plt.bar(['not canceled','canceled'],df['is_canceled'].value_counts() ,edgecolor = 'g',width= 0.7)
plt.show()
```

Bar graph shows cancelation and not calation rate of hotels.

```python
#comparisson of city hotel nresort hotel cancelation
plt.figure(figsize=(8,4))
ax1= sns.countplot(x ='hotel',hue='is_canceled',palette=('red','yellow'), data=df)
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend('bbox_to_anchor',(1,1))
plt.title('reservation status of different hotels',size=20)
plt.xlabel('hotels')
plt.ylabel('number of reservation')
plt.legend(['not canceled','canceled'])
plt.show()
```

The bar graph shows the percentage of reservation that are canceled and thoes that are not_canceled.in comparison to resort hotels,city hotels have more bookings.

```python
# for to finding resort hotel the how many percent of reservations are cancelled and how many percent of reservation are not cancelled.

resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#for to finding city hotel the how many percent of reservations are cancelled and how many percent of reservation are not cancelled.

city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title("Average Daily Rate in City and Resort Hotel",fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
#we can see in which month maximum reservation and in which month maximum cancellation.
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='month',hue='is_canceled',data=df,palette='bright')
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend('bbox_to_anchor',(1,1))

plt.title('reservation status per month',size=40)
plt.xlabel('month')
plt.ylabel('number of reservation')
plt.legend(['not canceled','canceled'])
plt.show()

```

```python
plt.figure(figsize = (8,4))
plt.title('ADR per month',fontsize=10)
sns.barplot(x='month' ,y= 'adr' , data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())

plt.show()
```

This grouped bar graph to analyze the months with the highest and lowest reservation rate according to reservation status.number of canceled reservations are largest in the month of august. whereas january is the month with the most canceled reservations.

```python
#for  to identify which contry has high reservation canceled

canceled_data= df[df['is_canceled']==1]
top_10_country=canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(15,8))
plt.title('top 10 countries of high reservation canceled')
plt.pie(top_10_country,autopct='%.2f',labels=top_10_country.index)
plt.show()
```

Pie chart shows top 10 countries of reservation cancelation.

```python
canceled_df_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace = True)
canceled_df_adr.sort_values('reservation_status_date',inplace=True)

not_canceled_data = df[df['is_canceled'] == 0]
not_canceled_df_adr=not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_df_adr.reset_index(inplace = True)
not_canceled_df_adr.sort_values('reservation_status_date',inplace = True)

plt.figure(figsize = (20,6))
plt.title('Average daily rate')
plt.plot(not_canceled_df_adr['reservation_status_date'],not_canceled_df_adr['adr'],label = 'not cancelled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'],label= 'canceled')
plt.legend()
```
